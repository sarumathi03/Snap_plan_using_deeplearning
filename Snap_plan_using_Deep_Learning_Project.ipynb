EXTRACTION OF  DATE FROM A IMAGE IN A CROPPED FORMAT

```python
img = cv2.imread('/content/drive/My Drive/Snap plan/Event details/Webinar on Machine Learning - Broucher.jpeg')
d = pytesseract.image_to_data(img, output_type=Output.DICT)
keys = list(d.keys())

date_pattern = '^(0[1-9]|[12][0-9]|3[01]).(0[1-9]|1[012]).(19|20)\d\d$' or '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
#phone_pattern='^[0-9][0-9][0-9]-[0-9][0-9][0-9]-[0-9][0-9][0-9]$'
#amount_pattern='^[1][1][1]-[2][2][2]-[3][3][3],$'

n_boxes = len(d['text'])
for i in range(n_boxes):
    if int(d['conf'][i]) > 60:
    	if re.match(date_pattern, d['text'][i]):
	        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
	        #img = cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

img1 = Image.fromarray(img, 'RGB')
from PIL import Image 
  
# Setting the points for cropped image 
left = x
top = y
right = x+w
bottom = y+h
  
# Cropped image of above dimension 
# (It will not change orginal image) 
im1 = img1.crop((left, top, right, bottom)) 
  
# Shows the image in image viewer 
display(im1)
im1.save('/content/drive/My Drive/Snap plan/Output files/cropped.jpg')
```

EXTRACTING DATE, TIME AND LINK 


```python
import re

file=open('/content/drive/My Drive/Snap plan/extracted.txt','w')
file.write(txt)
file.close()



fopen=open('/content/drive/My Drive/Snap plan/extracted.txt','r')
dformat='(0[1-9]|[12][0-9]|3[01]).(0[1-9]|1[012]).(19|20)\d\d$'
tformat='(([01]\d|2[0-3]):([0-5]\d)|24:00)$'
lformat='^(http:\/\/|https:\/\/www\.|http:\/\/www\.|https:\/\/)?[a-z0-9]+([\-\.]{1}[a-z0-9]+)*\.[a-z]{2,5}(:[0-9]{1,5})?(\/.*)?$'



for i in fopen.readlines():
  for a in i.split():
    #print(a)
    if(re.match(dformat,a)):
      date=a
    if(re.match(tformat,a)):
      time=a
    if(re.match(lformat,a)):
      link=a

print(date)
print(time)
print(link)
```

WRITING THE EXTRACTED DATE, TIME AND LINK TO A TEXT FILE

```python
with open('/content/drive/My Drive/Snap plan/Output files/For voice.txt', 'w') as file:
  file.write("You have an event scheduled on ")
  file.write(date)
  file.write(" at ")
  file.write(time)
  file.write(". The link to join is ")
  for i in link:
    file.write(i)
  file.close()

with open('/content/drive/My Drive/Snap plan/Output files/For voice.txt', 'r') as file:
    data = file.read().replace('\n', '')
    file.close()
print(data)
```

PARSING THE DATE EXTRACTED INTO DATE, MONTH AND YEAR

```python
import datetime
dat=datetime.datetime.strptime(date,"%d.%m.%Y")
date=dat.day
print(date)
```

```python
import calendar 
  
yy = dat.year
mm = dat.month
  

# display the calendar 
print(calendar.month(yy, mm))
```
